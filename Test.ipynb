```python
from ultralytics import YOLO
import cv2
```

```python
# Create a new YOLO model from scratch
model = YOLO("yolo11n.yaml")

# Load a pretrained YOLO model (recommended for training)
model = YOLO("yolo11n.pt")

# Train the model using the 'coco8.yaml' dataset for 3 epochs
results = model.train(data="coco8.yaml", epochs=3)
```

```python
# Evaluate the model's performance on the validation set
results = model.val()
```

```python
# Perform object detection on an image using the model
results = model("KITTI_Selection/images/006310.png")
```

```python
# Export the model to ONNX format
success = model.export(format="onnx")
```

```python
for result in results:
    # .plot() draws the boxes on the image and returns a numpy array (BGR)
    img_with_boxes = result.plot()

    # Save the result to a file
    cv2.imwrite('output_detected.jpg', img_with_boxes)
    
    # Optional: Display it in a window (press any key to close)
    cv2.imshow('YOLO Detection', img_with_boxes)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```
